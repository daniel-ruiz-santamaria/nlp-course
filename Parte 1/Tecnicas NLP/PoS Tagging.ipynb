Ahora vamos a obtener del texto solo un tipo de parlabras, pasado como parámetro


```python
def filtrar_palabras_por_tipo(document, tipo):
  filtered = []
  for token in document:
      if token.pos_ == tipo:
          filtered.append(token.lemma_)
  counter = Counter(filtered)
  return counter.most_common()
```

```python
# Obtener Adjetivos
print(filtrar_palabras_por_tipo(document, 'ADJ'))
```

```python
# Obtener Adjetivos
print(filtrar_palabras_por_tipo(document, 'VERB'))
```

```python
# Obtener Verbos
print(filtrar_palabras_por_tipo(document, 'NOUN'))
```

Tambien podemos filtrar el texto, para obtener las frases donde aparece una determinada palabra clave

```python
import re
from IPython.display import Markdown, display
```

```python
import re
from IPython.display import Markdown, display
def find_sentences_with_keyword(keyword, document):
    
    #Recorremos todas las frases del texto
    for sentence in document.sents:
        sentence = sentence.text
        
        if keyword.lower() in sentence.lower(): # Comporbamos si la palabra en minuscula se encuentra en la frase en minuscula
            
            sentence = re.sub('\n', ' ', sentence) # Usamos expresiones regulares para sustituir saltos de linea por espacios
            sentence = re.sub(f"{keyword}", f"**{keyword}**", sentence, flags=re.IGNORECASE) # Sustituimos la palabra clave, por la palabra clave rodeada por dos asteriscos, que es la sintaxis markdown para las palabras en negrita
            
            display(Markdown(sentence)) # Mostramos el texto en formato Markdown
```

```python
find_sentences_with_keyword(keyword="pueblo", document=document)
```

En ocasiones tambien es interesante obtener las palabras de contexto, situadas a la izquierda de la palabra que buscamos.

En este caso vamos a crear la funcion ngramas, que nos muestra la secuencia de n palabras en el texto

```python
# Empezamos por crear una lista de tuplas, de las palabras y sus tipos 
tokens_and_labels = [(token.text, token.pos_) for token in document if token.is_alpha]
```

```python
#Make a function to get all two-word combinations
def get_ngrams(document, number_consecutive_words=2):
    # Empezamos por crear una lista de tuplas, de las palabras y sus tipos 
    tokens_and_labels = [(token.text, token.pos_) for token in document if token.is_alpha]
    ngrams = [] # Inicializamos la lista de ngramas
    length_of_word_list = len(tokens_and_labels) - (number_consecutive_words - 1) # Ajustamos el tamaño , a el tamaño de palabras en el documento - numero de palabras consecutivas -1
    
    # Iteranos netre 0 y la longitud calculada length_of_word_list
    for word_index in range(length_of_word_list):
        
        #Index the list at each number, grabbing the word at that number index as well as N number of words after it
        ngram = tokens_and_labels[word_index : word_index + number_consecutive_words]
        
        #Append this word combo to the master list "ngrams"
        ngrams.append(ngram)
        
    return ngrams
```

```python
ngrams  = get_ngrams(document=document, number_consecutive_words=3)
```

Ahora vamos a obtener las palabras que mas frecuentemente aparecen en el texto, junto a la palabra que pasemos por parámetro

```python
def obtener_palabras_vecinias(keyword, ngramas, pos_label = None):
    
    neighbor_words = []
    keyword = keyword.lower()
    
    for ngram in ngramas:
      
        # Extraemos en minusculas las palabras de los ngramas, y las guardamos en la lista words
        words = [word.lower() for word, label in ngram]        
        
        #Comprebamos si palabra clave esta en el ngrama
        if keyword in words:
            
            for word, label in ngram: # Si lo esta, recorremos todas las palabras del ngrama
                
                # Por cada palabra, comprobamos si no coincide con la palabra clave, y en ese caso es vecina
                if word.lower() != keyword:
                    # Comprobamos si el tipo de la palabra coincide con el tipo pasado como parámetro, o si el tipo pasado es None, y de ser así añadimos a la lista
                    if label == pos_label or pos_label == None:
                        neighbor_words.append(word.lower())
    
    return Counter(neighbor_words).most_common() # Contamos las apariciones en la lista de las palabras, y las ordenamos por frecuencia
```

```python
# Buscamos las palabras vecinas a bien, y mostramos las frcuencias
obtener_palabras_vecinias("bien",ngrams)
```

```python
# Igual que el anterior, pero filtrando por verbos
obtener_palabras_vecinias("bien", ngrams, pos_label='VERB')
```
