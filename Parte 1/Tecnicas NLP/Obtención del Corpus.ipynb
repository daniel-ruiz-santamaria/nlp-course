* Obtención de un html desde una URL

En este caso, la lectura es como cualquier otra lectura desde una URL, es decir podemos hacerlo con la librería **requests** como en el ejemplo anterior

```python
URL = "https://decidesoluciones.es/procesamiento-del-lenguaje-natural-pln-o-nlp-que-es-y-para-que-se-utiliza/" # Pagina Web en formato HTML
```

```python
req = requests.get(URL)
html = req.text
print(html)
```

En este caso, lo que obtenemos, es la pagina Web en formato HTML.

Podemos usar la librería **beautifulsoup4**, para extraer de forma sencilla el texto que aparece en el documento HTML

![](https://stackabuse.s3.amazonaws.com/media/parsing-html-with-beautifulsoup-in-python-1.jpg)

Para ello la instalamos

```python
!pip install beautifulsoup4 # Instalación de la librería
```

Y la importamos

```python
from bs4 import BeautifulSoup
```

Y creamos una instancia del objeto BeautifulSoup, pasandole como parámetro el html

```python
soup = BeautifulSoup(html, "html5lib") # Crear instancia del objeto BeautifulSoup, a partir del texto en html, interpretado en formato html5
```

A partir de aqui ya podemos trabajar con la instancia de **BeautifulSoup (soup)**

```python
print('titulo:', soup.head.title)
```

Encontrar todo el texto entre etiquetas **\<p>\</p>** (párrafos)

```python
for linea in soup.find_all("p"): # Extraemos el texto, entre etiquetas <p></p>
    print(linea.get_text())
```
