# Funciones utiles para trabajar con textos en python

## Funciones útiles para cadenas de texto

```python
texto = "Esto es un texto de prueba"
```

Funciones útiles

- Funciones para chequear

```python
# Chequea si la cadena de texto comienza con el substring pasado como parámetro
texto.startswith('E')
```

```python
# Chequea si la cadena de texto termina con el substring pasado como parámetro
texto.endswith('a')
```

```python
# Chequea si la subcadea A esta contenida en B
"de" in texto
```

```python
# Chequea si todos los caracteres estan en minusculas
texto.islower()
```

```python
# Chequea si todos los caracteres estan en mayúsculas
texto.isupper()
```

```python
# Chequea si todos los caracteres estan son alfabéticos
texto.isalpha() # Es falso por los espacios
```

```python
# Chequea si todos los caracteres estan son alfanuméricos
texto.isalnum() # Es falso por los espacios
```

```python
# Chequea si todos los caracteres son números
texto.isdigit() 
```

```python
# Chequea si cada palabra del texto empieza con una lera mayuscula y el resto son minúsculas
texto.istitle() 
```

- Funciones de Búsqueda

```python
# Retorna el indice de la primera aparición del la subcadena pasada como parametro, y -1 en caso de no estar presente
texto.find("e")
```

```python
# Retorna el indice de la última aparición del la subcadena pasada como parametro, y -1 en caso de no estar presente
texto.rfind("e")
```

```python
# Igual que find, solo que retorna una excepción si no lo encuentra
texto.index("e")
```

```python
# Igual que rfind, solo que retorna una excepción si no lo encuentra
texto.rindex("e")
```

```python
# Concatena los elementos de la lista pasada como parámetro, usando el la cadena para unir los elementos
"-".join(["esto","es","una","lista","unida","por","guiones"]) 
```

```python
# Divide una cadena de texto o una frase, cuando encuentra el caracter pasado por parámetro. Puede ser muy útil para dividir palabras por espacio o frases por puntos. Retorna una lista
texto.split(" ")
```

```python
# Divide una cadena de texto en frases al encontrar saltos de linea
"""Esto es un texto de prueba
Esto es otro
Y aqui el último""".splitlines()
```

```python
# Pone en mayuscula la primera letra de cada palabra
texto.title()
```

```python
# Quita espacios en blanco al principio y al final de un texto
"   Esto es un texto de prueba   ".strip()
```

```python
# Reemplaza la subcadena pasada como 1 parámetro en el texto, por la subcadena del 2 parámetro en el texto
texto.replace("prueba","ejemplo")
```
